```python
class Team:
    def __init__(self, id,name, abbreviation,nickname,city,state,year_founded):
        self.id=id
        self.name = name
        self.abbreviation = abbreviation
        self.nickname = nickname
        self.city = city
        self.state = state
        self.year_founded = year_founded

    def __str__(self) -> str:
        return f'{self.id} {self.name} {self.abbreviation} {self.nickname} {self.city} {self.state} {self.year_founded}'    
    
class Player:
    def __init__(self, id, full_name, first_name, last_name, is_active):
        self.id = id
        self.full_name = full_name
        self.first_name = first_name
        self.last_name = last_name
        self.is_active = is_active

    def __str__(self) :
        return f'{self.id} {self.full_name} {self.first_name} {self.last_name} {self.is_active}'
    
class PlayerCareerStats:
    def __init__(self, id,teamid,team, season,age,pts):
        self.id = id
        self.teamid = teamid
        self.team = team
        self.season = season
        self.age = age
        self.pts = pts

    def __str__(self) :
        return f'{self.id} {self.teamid} {self.team} {self.season} {self.age} {self.pts}'
```

```python
from nba_api.stats.static import teams
from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats

nba_teams = teams.get_teams()
nba_players = players.get_players()

```

```python
nbateams=[]

for team in nba_teams:
    nbateams.append(Team(id=team["id"],name=team["full_name"],abbreviation=team["abbreviation"],nickname=team["nickname"],city=team["city"],state=team["state"],year_founded=team["year_founded"]))

pd.DataFrame([t.__dict__ for t in nbateams ])
```

```python
nbaplayers=[]
for player in nba_players:
    nbaplayers.append(Player(id=player["id"],full_name=player["full_name"],first_name=player["first_name"],last_name=player["last_name"],is_active=player["is_active"]))

pd.DataFrame([t.__dict__ for t in nbaplayers ]).head()
```

```python
def getPlayer(playerName):
    for player in nbaplayers:
        if(player.full_name == playerName):
            return player
        
def getTeam(abbreviation):
    for team in nbateams:
        if(team.abbreviation == abbreviation):
            return team
        
def getPlayerStats(playerId):
    nbaplayerstats=[]
    career = playercareerstats.PlayerCareerStats(player_id=playerId)
    # print(career.get_available_data())
    # print(career.get_data_frames()[0])
    # print(career.get_data_frames()[1])
    seasonTotalsList=career.get_normalized_dict()['SeasonTotalsRegularSeason']
    for seasonTotals in seasonTotalsList:
        nbaplayerstats.append(PlayerCareerStats(id=seasonTotals['PLAYER_ID'],teamid=seasonTotals['TEAM_ID'],team=seasonTotals['TEAM_ABBREVIATION'],season=seasonTotals['SEASON_ID'],age=seasonTotals['PLAYER_AGE'],pts=seasonTotals['PTS'])    )
    return nbaplayerstats
```

```python
lebron = getPlayer("LeBron James")
koby = getPlayer("Kobe Bryant")
jordan = getPlayer("Michael Jordan")
jalen = getPlayer("Jalen Brunson")

print(lebron)
print(koby)
print(jordan)
print(jalen)
```

```python
class PlayerSummary:
    def __init__(self, team, totalpoints, totalseasons):
        self.team = team
        self.totalpoints = totalpoints
        self.totalseasons = totalseasons
        self.avgpoints = 0
    
    def calcAvgPoints(self):
        self.avgpoints = self.totalpoints / self.totalseasons

def playerSummary(playerid):
    nbaplayerstats = getPlayerStats(playerid)
    playerSummary = {}

    for season in nbaplayerstats:
        if(season.team in playerSummary):
            playerSummary[season.team].totalpoints += season.pts
            playerSummary[season.team].totalseasons += 1
        else:
            playerSummary[season.team]=PlayerSummary(season.team,season.pts,1)

    for team in playerSummary:
        playerSummary[team].calcAvgPoints()

    return playerSummary
            
```
